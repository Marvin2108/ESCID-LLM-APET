```python
import matplotlib.pyplot as plt

def plot_precision_recall_curve(precision, recall, class_name):
    """
    Stellt die Precision-Recall-Kurve fr eine Klasse dar.
    
    :param precision: Liste oder Array mit den Precision-Werten
    :param recall: Liste oder Array mit den Recall-Werten
    :param class_name: Name der Klasse (String) fr die Beschriftung
    """
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, marker='.', label=f'Precision-Recall Curve ({class_name})')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall-Kurve fr Klasse: {class_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Beispiel: Konkrete Werte fr Precision und Recall
precision_negativ = [0.83, 0.81] 
recall_negativ = [0.37, 0.51]

# Aufruf der Funktion fr die Klasse "negativ"
plot_precision_recall_curve(precision_negativ, recall_negativ, "neutral")
```

```python
import matplotlib.pyplot as plt

def plot_precision_recall_curves(precision_baseline, recall_baseline, 
                                 precision_oneshot, recall_oneshot, class_name):
    """
    Stellt zwei Precision-Recall-Kurven (Baseline und One-Shot) fr eine Klasse dar.
    
    :param precision_baseline: Liste oder Array mit Precision-Werten fr den Baseline-Ansatz
    :param recall_baseline: Liste oder Array mit Recall-Werten fr den Baseline-Ansatz
    :param precision_oneshot: Liste oder Array mit Precision-Werten fr den One-Shot-Ansatz
    :param recall_oneshot: Liste oder Array mit Recall-Werten fr den One-Shot-Ansatz
    :param class_name: Name der Klasse (String) fr die Beschriftung
    """
    plt.figure(figsize=(8, 6))
    
    # Baseline-Kurve
    plt.plot(recall_baseline, precision_baseline, marker='o', linestyle='-', color='blue', label='Baseline')
    
    # One-Shot-Kurve
    plt.plot(recall_oneshot, precision_oneshot, marker='s', linestyle='--', color='green', label='One-Shot')
    
    # Diagramm-Details
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall-Kurven fr Klasse: {class_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Beispiel: Konkrete Werte fr Precision und Recall
precision_baseline = [0.83] 
recall_baseline = [0.37]

precision_oneshot = [0.81]
recall_oneshot = [0.51]

# Aufruf der Funktion fr die Klasse "negativ"
plot_precision_recall_curves(precision_baseline, recall_baseline, 
                             precision_oneshot, recall_oneshot, "negativ")
```
